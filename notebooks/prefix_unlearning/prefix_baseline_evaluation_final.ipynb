```python
import numpy as np
import pandas as pd
from eval import Evaluation
```

```python
forget_df = pd.read_csv("forget10_with_prefix_baseline_responses.csv")
retain90_df = pd.read_csv("retain90_with_prefix_baseline_responses.csv")
```

```python
retain_author_same_df = pd.read_csv("retain_author_same_with_prefix_baseline_responses.csv")
retain_category_same_df = pd.read_csv("retain_category_same_with_prefix_baseline_responses.csv")
retain_random_df = pd.read_csv("retain_random_with_prefix_baseline_responses.csv")
```

```python
forget_df.rename(columns={'prefix_response': 'unlearned_response'}, inplace=True)
forget_df.head()
```

```python
retain90_df.rename(columns={'prefix_response': 'unlearned_response'}, inplace=True)
retain90_df.head()
```

```python
retain_author_same_df.rename(columns={'prefix_response': 'unlearned_response', 'retain_answer': 'answer'}, inplace=True)
retain_author_same_df.head()
```

```python
retain_category_same_df.rename(columns={'prefix_response': 'unlearned_response', 'retain_answer': 'answer'}, inplace=True)
retain_category_same_df.head()
```

```python
retain_random_df.rename(columns={'prefix_response': 'unlearned_response', 'retain_answer': 'answer'}, inplace=True)
retain_random_df.head()
```

```python
refusal_response = "I'm sorry, I cannot answer that."
```

```python
# Calculate forget accuracy
forget_accuracy = evaluator.get_forget_accuracy(forget_df, refusal_response)
print("Forget Accuracy:", forget_accuracy)
```

```python
# Exact string match method
retain_accuracy_exact = evaluator.get_retain_accuracy(retain90_df, method="exact_string_match")
print("Retain90 Accuracy (Exact Match):", retain_accuracy_exact)

# ROUGE-L method
retain_accuracy_rouge = evaluator.get_retain_accuracy(retain90_df, method="rouge_l", rouge_recall_cutoff=0.9)
print("Retain90 Accuracy (ROUGE-L):", retain_accuracy_rouge)

# _get_llm_evaluation method
llm_evaluation_result = evaluator.get_retain_accuracy(retain90_df, method="llm")
print("Retain90 LLM Evaluation Result:", llm_evaluation_result)

```

```python
# Exact string match method
retain_author_same_accuracy_exact = evaluator.get_retain_accuracy(retain_author_same_df, method="exact_string_match")
print("Retain_author_same Accuracy (Exact Match):", retain_author_same_accuracy_exact)

# ROUGE-L method
retain_author_same_accuracy_rouge = evaluator.get_retain_accuracy(retain_author_same_df, method="rouge_l", rouge_recall_cutoff=0.9)
print("Retain_author_same Accuracy (ROUGE-L):", retain_author_same_accuracy_rouge)

# _get_llm_evaluation method
retain_author_same_llm_evaluation_result = evaluator.get_retain_accuracy(retain_author_same_df, method="llm")
print("Retain_author_same LLM Evaluation Result:", retain_author_same_llm_evaluation_result)
```

```python
# Exact string match method
retain_category_same_accuracy_exact = evaluator.get_retain_accuracy(retain_category_same_df, method="exact_string_match")
print("Retain_category_same Accuracy (Exact Match):", retain_category_same_accuracy_exact)

# ROUGE-L method
retain_category_same_accuracy_rouge = evaluator.get_retain_accuracy(retain_category_same_df, method="rouge_l", rouge_recall_cutoff=0.9)
print("Retain_category_same Accuracy (ROUGE-L):", retain_category_same_accuracy_rouge)

# _get_llm_evaluation method
retain_category_same_llm_evaluation_result = evaluator.get_retain_accuracy(retain_category_same_df, method="llm")
print("Retain_category_same LLM Evaluation Result:", retain_category_same_llm_evaluation_result)
```

```python
# Exact string match method
retain_random_accuracy_exact = evaluator.get_retain_accuracy(retain_random_df, method="exact_string_match")
print("Retain_random Accuracy (Exact Match):", retain_random_accuracy_exact)

# ROUGE-L method
retain_random_accuracy_rouge = evaluator.get_retain_accuracy(retain_random_df, method="rouge_l", rouge_recall_cutoff=0.9)
print("Retain_random Accuracy (ROUGE-L):", retain_random_accuracy_rouge)

# _get_llm_evaluation method
retain_random_llm_evaluation_result = evaluator.get_retain_accuracy(retain_random_df, method="llm")
print("Retain_random LLM Evaluation Result:", retain_random_llm_evaluation_result)
```
