```python
import os
import sys

current_dir = os.getcwd()
root_dir = os.path.abspath(os.path.join(current_dir, '..', '..'))
sys.path.append(root_dir)
```

```python
import numpy as np
import pandas as pd
from eval import Evaluation
```

```python
evaluator = Evaluation()
```

```python
forget_df = pd.read_csv("forget10_with_prefix_responses.csv")
retain_df = pd.read_csv("retain90_with_prefix_responses.csv")
```

```python
forget_df.rename(columns={'prefix_response': 'unlearned_response'}, inplace=True)
forget_df.head()
```

```python
retain_df.rename(columns={'prefix_response': 'unlearned_response'}, inplace=True)
retain_df.head()
```

```python
refusal_response = "I'm sorry, I cannot answer that."
```

```python
# Calculate forget accuracy
forget_accuracy = evaluator.get_forget_accuracy(forget_df, refusal_response)
print("Forget Accuracy:", forget_accuracy)
```

```python
# Exact string match method
retain_accuracy_exact = evaluator.get_retain_accuracy(retain_df, method="exact_string_match")
print("Retain Accuracy (Exact Match):", retain_accuracy_exact)

# ROUGE-L method
retain_accuracy_rouge = evaluator.get_retain_accuracy(retain_df, method="rouge_l", rouge_recall_cutoff=0.9)
print("Retain Accuracy (ROUGE-L):", retain_accuracy_rouge)
```
